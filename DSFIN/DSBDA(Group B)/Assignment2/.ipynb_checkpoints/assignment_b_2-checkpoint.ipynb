# Data Integration

```python
df1.shape
```

```python
df2.shape
```

```python
min_len = min(len(df1), len(df2))
final_df = pd.concat([df2.iloc[:min_len,:], df1.iloc[:min_len,:]], axis=1)
```

# Model 

```python
#X = final_df.drop(["HeartDisease"] , axis =1 )
X = final_df.drop("target", axis=1)
y = final_df["target"]
```

```python
y = final_df.target
```

```python
X
```

```python
y
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2, random_state=7)

model = LogisticRegression()
model.fit(Xtrain, ytrain)
```

```python
print(final_df.dtypes)
```

```python
print(final_df.isnull().sum())
```

```python
final_pred=model.predict(Xtest)
final_pred
```

```python
pd.DataFrame({"Actual":ytest, "Predict":final_pred})
```

```python
print("Accuracy of the Logistics Regression Model is ::",accuracy_score(ytest, final_pred)*100, "%")
```
