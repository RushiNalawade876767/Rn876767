```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
print(df.head())
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(), cmap='YlGnBu', annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.lineplot(data=df, x='age', y='cp', hue='target')
plt.title("Age vs CP")
plt.show()
```

```python
sns.lineplot(data=df, x='age', y='trestbps', hue='target')
plt.title("Age vs RestBP")
plt.show()
```

```python
sns.lineplot(data=df, x='age', y='chol', hue='target')
plt.title("Age vs Cholesterol")
plt.show()
```

```python
sns.lineplot(data=df, x='oldpeak', y='exang', hue='target')
plt.title("Oldpeak vs Exercise Induced Angina")
plt.show()
```

```python
sns.histplot(data=df, x='target', hue='sex')
plt.title("Heart Disease Distribution by Sex")
plt.show()
```

```python
sns.histplot(data=df, x='cp', hue='sex')
plt.title("CP Types Distribution by Sex")
plt.show()
```

```python
sns.histplot(data=df, x='cp', hue='target')
plt.title("CP vs Heart Disease")
plt.show()
```

```python
sns.histplot(data=df, x='age', hue='target')
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=df, x='trestbps', hue='target')
plt.title("RestBP Distribution")
plt.show()
```

```python
temp_df = df[['age','cp','trestbps','chol','fbs','oldpeak','target']]

sns.pairplot(temp_df, hue='target')
plt.suptitle("Looking for Insights in Data", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(temp_df.columns, 1):
    plt.subplot(4,3,i)
    plt.title(f"Distribution of {col} Data")
    sns.histplot(temp_df[col], kde=True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['target'].value_counts().plot.pie(
    autopct='%1.1f%%',
    labels=['No Disease','Disease'],
    colors=['#66b3ff','#ff9999']
)
plt.title("Heart Disease Distribution")
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['chol'], alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Scatter Plot: Age vs Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='target', y='chol', data=df)
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol")
plt.title("Box Plot: Cholesterol vs Heart Disease")
plt.show()
```
