```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AirQuality.csv', delimiter=';')
df
```

```python
df = df.rename(columns={'T': 'Temperature'})
df = df.rename(columns={'RH': 'Relative Humidity'})
df = df.rename(columns={'AH': 'Absolute Humidity'})
df
```

```python
df['CO(GT)'] = df['CO(GT)'].str.replace(',', '.').astype(float)
df['C6H6(GT)'] = df['C6H6(GT)'].str.replace(',', '.').astype(float)
df['Temperature'] = df['Temperature'].str.replace(',', '.').astype(float)
df['Relative Humidity'] = df['Relative Humidity'].str.replace(',', '.').astype(float)
df['Absolute Humidity'] = df['Absolute Humidity'].str.replace(',', '.').astype(float)
df
```

```python
df = df.drop_duplicates()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR
    return column[(column >= Q1 - threshold) & (column <= Q3 + threshold)]

col_name = ['Temperature','Relative Humidity','Absolute Humidity','PT08.S4(NO2)','PT08.S5(O3)','C6H6(GT)','PT08.S2(NMHC)','PT08.S1(CO)']

for col in col_name:
    df[col] = remove_outliers(df[col])
```

```python
df.loc[:, 'Year'] = pd.to_datetime(df['Date'], dayfirst=True).dt.year
df.loc[:, 'month'] = pd.to_datetime(df['Date'], dayfirst=True).dt.month

df.loc[:, 'Year'] = df['Year'].astype(str)
df.loc[:, 'month'] = df['month'].astype(str)
```
