```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AirQuality.csv', delimiter=';')
df
```

```python
df = df.rename(columns={'T': 'Temperature'})
df = df.rename(columns={'RH': 'Relative Humidity'})
df = df.rename(columns={'AH': 'Absolute Humidity'})
df
```

```python
df['CO(GT)'] = df['CO(GT)'].str.replace(',', '.').astype(float)
df['C6H6(GT)'] = df['C6H6(GT)'].str.replace(',', '.').astype(float)
df['Temperature'] = df['Temperature'].str.replace(',', '.').astype(float)
df['Relative Humidity'] = df['Relative Humidity'].str.replace(',', '.').astype(float)
df['Absolute Humidity'] = df['Absolute Humidity'].str.replace(',', '.').astype(float)
df
```

```python
df = df.drop_duplicates()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def remove_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    threshold = 1.5 * IQR

    lower = Q1 - threshold
    upper = Q3 + threshold

    return df[(df[column] >= lower) & (df[column] <= upper)]

col_name = [
    'Temperature',
    'Relative Humidity',
    'Absolute Humidity',
    'PT08.S4(NO2)',
    'PT08.S5(O3)',
    'C6H6(GT)',
    'PT08.S2(NMHC)',
    'PT08.S1(CO)'
]

for col in col_name:
    df = remove_outliers(df, col)

df = df.copy()
```

```python
df['Year'] = pd.to_datetime(df['Date'], dayfirst=True).dt.year.astype(str)
df['month'] = pd.to_datetime(df['Date'], dayfirst=True).dt.month.astype(str)
```
